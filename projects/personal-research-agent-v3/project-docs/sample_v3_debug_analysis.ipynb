```
import matplotlib.pyplot as plt
import pandas as pd

reason_counts = {}
selected_counts = {
  "bitcoin": 1,
  "events": 1,
  "news": 1
}
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
reason_series = pd.to_numeric(pd.Series(reason_counts), errors='coerce').dropna()
selected_series = pd.to_numeric(pd.Series(selected_counts), errors='coerce').dropna()

if reason_series.empty:
    axes[0].set_title('Rejection reasons')
    axes[0].text(0.5, 0.5, 'No rejection reasons', ha='center', va='center')
    axes[0].set_xlabel('Count')
    axes[0].set_yticks([])
else:
    reason_series.sort_values().plot(kind='barh', ax=axes[0], title='Rejection reasons')
    axes[0].set_xlabel('Count')

if selected_series.empty:
    axes[1].set_title('Selected counts')
    axes[1].text(0.5, 0.5, 'No selected items', ha='center', va='center')
    axes[1].set_xlabel('Count')
    axes[1].set_yticks([])
else:
    selected_series.sort_values().plot(kind='barh', ax=axes[1], title='Selected counts')
    axes[1].set_xlabel('Count')
plt.tight_layout()
```
